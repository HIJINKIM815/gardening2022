**시험에 맞는 데이터 셋 만들기 , 칼럼 =['나라 이름', '날짜', '누적 확진자수']**

```python
# 데이터 불러오기 
df = pd.read_csv('./data/covid.csv')
```

```python
df.head(10)
```

```python
# 필요한 데이터만 가져오기 ('국가명','날짜','확진자수')
df = df[['iso_code','date','total_cases']]
# 새로운 칼럼명으로 지정
df.columns = ['coun','date','confirmed']
df
```

```python
# 가장 최근 날짜를 기준으로 가장 빈도수가 많은 20개국을 뽑기
df_12_19 = df[df['date']=='2021-12-19']
df_12_19 = df_12_19.sort_values('confirmed',ascending=False)
```

```python
df_12_19['coun'][0:30].values===[[[[[[[[[[]]]]]]]]]]
```

```python
coun = ['USA', 'KOR', 'IND','BRA', 'GBR', 'RUS', 'TUR', 'FRA', 'DEU', 'IRN', 'ESP',
       'ARG', 'ITA', 'COL', 'IDN', 'POL', 'MEX', 'UKR', 'ZAF', 'NLD','PHL']

# 빈도수가 많은 20개국 나라만 데이터 프레임에서 뽑아내기 
df = df.loc[df['coun'].isin(coun)]
df
```

```python
df[df['coun']=='ARG'].isna().sum()
```

```python
# 시간 데이터로 변환
df['date']=pd.to_datetime(df['date'])
df.info()
```

```python
# 올해 2월 1일 이후부터 데이터만 뽑기
df = df[df['date']>='2021-02-01']
df = df.reset_index() # 인덱스 재 설정
df = df.drop('index',axis=1)
```

```python
# 완성
df.to_csv('./data/covid19.csv',index=False)
```

**나라별로 동일하게 322개의 데이터를 가지고 있다. ( 실제 시험에서도 각 나라별로 동일한 데이터 개수를 가지고 있었음 )**

```python
countries = df['coun'].value_counts().keys()
countries.tolist()
```

```python
groups = df.groupby(df.coun)
df1 = groups.get_group("ARG")['confirmed']
for country in countries:
    if country =='ARG':
        pass
    else:
        df1 = pd.concat([df1,groups.get_group(country)['confirmed']],axis=1)
df1.columns = countries
```
