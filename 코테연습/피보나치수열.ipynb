```python
n번째 피보나치 수를 구하시오.

피보나치 수열이란, 앞에 있는 두개의 수를 더해서 계속 해서 생성 되는 수열을 의미한다. 
첫번째와 두번째 수가 각각 1라면, 세번째 수는 2(1+1), 네번째 수는 3(1+2), 다섯번째 수는 5(2+3) ...이 된다.

```

image.png

```python
def fibonacci(n):
    fibonacci_list = [0,1,1]
    for i in range(3,n):
      fibonacci_list.append(fibonacci_list[i-2]+fibonacci_list[i-1])
    return fibonacci_list[n-1] 
```

```python
fibonacci(4)
```

```python
fibonacci(5)
```
